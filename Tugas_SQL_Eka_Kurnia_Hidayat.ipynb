### 1. Tampilkan data order yang mempunyai order amount dibawah 1000

```python
# Connect ke Database
connection = sqlite3.connect('data_penjualan.db')

# Buat Query SQL
result  = connection.cursor()
command = "SELECT * FROM orders WHERE ORD_AMOUNT < '1000.00';"

# Eksekusi Query SQL
result.execute(command)

# Hasil dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

### 4. Tampilkan informasi customer yang mempunyai payment amount diatas 8000 dan urutkan berdasarkan nama customer dari A sampai Z

```python
# Connect ke Database
connection = sqlite3.connect('data_penjualan.db')

# Buat Query SQL
result  = connection.cursor()
command = "SELECT * FROM customer WHERE PAYMENT_AMT > '8000.00' ORDER BY CUST_NAME ASC ;"

# Eksekusi Query SQL
result.execute(command)

# Hasil dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```

### 5. Tampilkan data customer yang tidak tinggal di USA dan memiliki grade lebih dari 1

```python
# Connect ke Database
connection = sqlite3.connect('data_penjualan.db')

# Query SQL
result  = connection.cursor()
command = "SELECT * FROM customer WHERE CUST_COUNTRY NOT IN ('USA') AND GRADE > '1';"

# Eksekusi Query SQL
result.execute(command)

# Hasil dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```
